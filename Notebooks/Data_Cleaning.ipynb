## Step 9: Convert Date Columns

```python
df["Date of Admission"] = pd.to_datetime(df["Date of Admission"])
df["Discharge Date"] = pd.to_datetime(df["Discharge Date"])
```

```python
df.info()
```

## Step 10: Create Length of Stay

```python
df["Length of Stay"] = (
    df["Discharge Date"] -
    df["Date of Admission"]
).dt.days
```

## Step 11: Create Age Group

```python
bins = [0,18,35,50,65,100]

labels = [
    "0-18",
    "19-35",
    "36-50",
    "51-65",
    "65+"
]

df["Age Group"] = pd.cut(
    df["Age"],
    bins=bins,
    labels=labels
)
```

## Step 12: Extract Date Features

```python
df["Year"] = df["Date of Admission"].dt.year

df["Month"] = df["Date of Admission"].dt.month_name()

df["Quarter"] = df["Date of Admission"].dt.quarter
```
