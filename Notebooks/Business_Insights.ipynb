# Business Insights & Recommendations

## Step 1: Load the Clean Dataset

```python
import pandas as pd

df = pd.read_csv("H:\Hospital Performance & Patient Analytics\clean_healthcare_dataset.csv")
```

## Step 2: Executive Summary

###  Executive Summary

This report analyzes hospital operations using patient demographic, billing,
medical condition, insurance, and admission data.

The objective is to identify operational trends, financial performance,
patient demographics, and opportunities for improvement.

## Step 3: KPI Summary

```python
summary = {
    "Total Patients": len(df),
    "Average Age": round(df["Age"].mean(),1),
    "Total Billing": round(df["Billing Amount"].sum(),2),
    "Average Billing": round(df["Billing Amount"].mean(),2),
    "Average Stay": round(df["Length of Stay"].mean(),1)
}

summary
```

## Step 4: Top 10 Hospitals

```python
hospital = (
    df.groupby("Hospital")["Billing Amount"]
      .sum()
      .sort_values(ascending=False)
      .head(10)
)

hospital
```

### Insight
Top hospitals contribute a significant portion of the total billing revenue,
indicating that patient volume and treatment costs are concentrated among a
limited number of hospitals.

## Step 5: Medical Condition Analysis

```python
condition = (
    df["Medical Condition"]
      .value_counts()
)

condition
```

### Insight
A small number of medical conditions account for most patient admissions.
Targeted healthcare programs for these conditions could improve patient outcomes.

## Step 6: Insurance Analysis

```python
insurance = (
    df.groupby("Insurance Provider")["Billing Amount"]
      .sum()
      .sort_values(ascending=False)
)

insurance
```

### Insight
Insurance providers contribute differently to hospital revenue,
which can help management evaluate partnerships and reimbursement patterns.


## Step 7: Admission Type

```python
admission = (
    df["Admission Type"]
      .value_counts()
)

admission
```

### Insight
Emergency admissions represent a major share of total admissions,
highlighting the importance of emergency department resource planning.

## Step 8: Age Group Analysis

```python
age = (
    df.groupby("Age Group")
      .size()
)

age
```

### Insight
Middle-aged and senior patients represent a significant portion of the hospital population,
indicating where preventive care and treatment planning should be focused.
